##KNN usando Numpy

```python
import numpy as np
from math import sqrt
from collections import Counter
```

```python

def distancia_euclidiana(x1,x2):
  distancia = np.linalg.norm(x1-x2)
  return distancia
```

```python
def get_neighbors (treino,teste,nvisinhos):
  distancias = list()
  for train_row in treino:
    d = distancia_euclidiana(teste, train_row)
    distancias.append((train_row,d))
  d.sort(key=lambda tup:tup[1])
  visinhos = list()
  for i in range(nvisinhos):
    visinhos.append(distancias[i][0])
  return visinhos  
```

```python
def Xknn (x_teste, x_treino, y_treino, k):
  final_output = []
  for i in range(len(x_teste)):       ## Pega os índices dos dados de teste
    d = []
    votes = []
    for j in range(len(x_treino)):    ##Pega os índices dos dados do treino e compara cada um com o dado teste
      dist = distancia_euclidiana(x_treino[j] , x_teste[i])    #Calcula a distancia
      d.append([dist, j])             ##Registra aqui na lista a distancia e o indice do treino
    d.sort()                          #Ordena a lista e pega apenas até a distancia k
    d = d[0:k]
    for d, j in d:
      votes.append(y_treino[j])        #Pega da lista os valores um por um e coloca o valor de y correspondente do treino
    ans = Counter(votes).most_common(1)[0][0]  #Pega apenas o elemento que mais que aparece
    final_output.append(ans)          #Devolve pra nós o elemento em uma lista.
            
  return final_output

def score(y_teste,prev):
  cont = 0
  for i in range(len(prev)):
    if prev[i] == y_teste[i]:
      cont += 1
  return cont / len(y_teste)   #Calcula a acurácia. O quanto bateu com o esperado
```

```python
df=pd.read_csv('winequality.csv')


df.describe()
```

```python
df=df.drop(columns=['Unnamed: 0'])
df = df.drop_duplicates()
df.describe()
```

```python
from sklearn import preprocessing
df2=df.drop(columns=['wine_is_red','quality'])  
nrm_scaler = preprocessing.MinMaxScaler()

 
df_scaled = nrm_scaler.fit_transform(df2.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides'	,'free sulfur dioxide',	'total sulfur dioxide'	,'density',	'pH',	'sulphates',	'alcohol'])
df_scaled['wine_is_red'] = df['wine_is_red']
df_scaled['quality'] = df['quality']
df_scaled = df_scaled.dropna()
df_scaled.describe()

```

```python
X = df_scaled[['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides'	,'free sulfur dioxide',	'total sulfur dioxide'	,'density',	'pH',	'sulphates',	'alcohol','wine_is_red']]

y = df_scaled[['quality']]

features = ['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides'	,'free sulfur dioxide',	'total sulfur dioxide'	,'density',	'pH',	'sulphates',	'alcohol','wine_is_red']
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train ,y_test = train_test_split(X,y,random_state =0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=30)
```

```python
knn.fit(X_train, y_train)
```

```python
resultado = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Acurácia do modelo: ' , accuracy_score(y_test, resultado))
```
